    <h2>Python да переменныйлар (Variables)</h2>
    <p>Переменныйлар маълумотларни вақтинча сақлаш учун махсус контейнерлардир. Улар бизга маълумотларни сақлаш, ўзгартириш ва турли амалларни бажариш имконини беради. Масалан, бирон бир қийматни бир неча марта ишлатишимиз керак бўлганда, ҳар сафар қийматни қайта ёзиш ўрнига переменныйдан фойдаланиш мумкин.</p>
    <p>Переменныйлар дастурлашнинг асосий қисми бўлиб, маълумотларни динамик равишда бошқариш имконини беради. Улар қуйидаги ҳолларда қўлланилади:</p>
    <ul>
        <li>Қийматларни вақтинча сақлаш.</li>
        <li>Қийматларни ўзгартириш.</li>
        <li>Маълумотлар устида амалларни бажариш.</li>
    </ul>

<div class="card">
    <h2>Переменный номлари</h2>
    <p>Переменный номлари уларни бошқариш ва уларга мурожаат қилиш учун ишлатилади. Python да переменный номларини белгилашда қуйидаги қоидаларга риоя қилиш лозим:</p>
    <ul>
        <li>Номлар ҳарф ёки пастки чизиқча (<code>_</code>) билан бошланиши керак.</li>
        <li>Номлар ҳарф, рақам ёки пастки чизиқчадан иборат бўлиши мумкин.</li>
        <li>Номларда бўшлиқлар ёки махсус белгилар (<code>!, @, #, ва ҳ.к.</code>) бўлмаслиги керак.</li>
        <li>Номлар катта-кичик ҳарфларга сезгир; масалан, <code>age</code> ва <code>Age</code> турли переменныйлар ҳисобланади.</li>
    </ul>
    <p>Қуйидаги мисол Python да тўғри ва нотўғри переменный номлари ҳақида тушунча беради:</p>
    <pre>
    <code>
        # Тўғри номлар
        my_variable = 10
        age = 25
        _name = "John"
    </code>
    <code>
        # Нотўғри номлар
        2variable = 5  # рақам билан бошланган
        my-variable = 15  # дефис ишлатилган
        my variable = 20  # бўшлиқлар бор
    </code>
    </pre>
    <p>Тўғри номланган переменныйлар дастурда қулайлик ва тушунарлилик яратади. Шу сабабли, номларни мазмунли ва маълумотга мос танлаш тавсия этилади.</p>
