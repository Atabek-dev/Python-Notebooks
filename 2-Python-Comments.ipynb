<div class="card">
    <h2>Python да изоҳлар (Comments)</h2>
    <p>Изоҳлар кодингизни тушунишни осонлаштиради ва унинг ўқилишига ёрдам беради. Python да изоҳлар қуйидаги мақсадларда қўлланилади:</p>
    <ul>
        <li>Кодни тушунтириш.</li>
        <li>Кодни ўқишни осонлаштириш.</li>
        <li>Тест жараёнида коднинг айрим қисмларини бажарилишини олдини олиш.</li>
    </ul>
</div>

<div class="card">
    <h2>Изоҳ яратиш</h2>
    <p>Изоҳлар <code>#</code> белгиси билан бошланади ва Python уларни эътиборсиз қолдиради:</p>
    <pre><code># Бу изоҳ
print("Hello, World!")</code></pre>
    <p>Юқоридаги мисолда, биринчи қатор изоҳ бўлиб, иккинчи қатордаги код бажарилади.</p>
</div>

<div class="card">
    <h2>Қатор охиридаги изоҳлар</h2>
    <p>Изоҳлар қатор охирида ҳам жойлаштирилиши мумкин, бу ҳолда Python қаторнинг қолган қисмини эътиборсиз қолдиради:</p>
    <pre><code>print("Hello, World!") # Бу изоҳ</code></pre>
    <p>Бу усул коднинг муайян қисми ҳақида тушунтириш беришда фойдали.</p>
</div>

<div class="card">
    <h2>Коднинг бажарилишини олдини олиш</h2>
    <p>Изоҳлар коднинг айрим қисмларини вақтинча ўчириш учун ҳам қўлланилиши мумкин:</p>
    <pre><code># print("Hello, World!")
print("Cheers, Mate!")</code></pre>
    <p>Юқоридаги мисолда, биринчи қатор изоҳланган ва бажарилмайди, иккинчи қатор эса бажарилади.</p>
</div>

<div class="card">
    <h2>Кўп қаторли изоҳлар</h2>
    <p>Python да кўп қаторли изоҳлар учун махсус синтаксис мавжуд эмас. Бир неча қаторли изоҳ яратиш учун ҳар бир қаторни <code>#</code> билан бошлаш мумкин:</p>
    <pre><code># Бу изоҳ
# бир нечта
# қаторда ёзилган
print("Hello, World!")</code></pre>
    <p>Ёки учтлаик кўштироқлардан фойдаланиш мумкин:</p>
    <pre><code>"""
Бу изоҳ
бир нечта қаторда
ёзилган
"""
print("Hello, World!")</code></pre>
    <p>Бу усул кўп қаторли изоҳлар ёзишда қулай.</p>

</div>
