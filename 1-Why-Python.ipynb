<div class="card">
    <h2>Python тарихига назар</h2>
    <p>Python дастурлаш тили 1991 йилда Guido van Rossum томонидан ишлаб чиқилган. Унинг асосий мақсади дастурчиларга содда, ўқиладиган ва интуитив код ёзиш имконини бериш эди. Python номи Monty Python юмор гуруҳининг номи билан боғлиқ.</p>
</div>

<div class="card">
    <h2>Python нинг афзалликлари</h2>
    <ul>
        <li><strong>Қулайлик ва соддалик</strong>: Python синтаксисини ўрганиш ва тушуниш жуда осон, бу дастурчиларга тезкор дастурлар яратиш имконини беради.</li>
        <li><strong>Кросс-платформалик</strong>: Python Windows, MacOS, Linux каби турли операцион тизимларда ишлайди.</li>
        <li><strong>Бой китобхоналар (libraries)</strong>: Python кенг кўламли стандарт ва учинчи томон китобхоналарига эга, улар орқали маълумотлар таҳлили, веб-ишланма, машиналарни ўрганиш ва бошқа соҳаларда ишлаш мумкин.</li>
        <li><strong>Катта ҳамжамият</strong>: Python дастурчилар жамоаси жуда фаол ва катта, бу эса янги дастурчиларга катта ёрдам беради.</li>
        <li><strong>Очиқ кодли (Open Source)</strong>: Python бутунлай бепул ва унинг кодини ҳамма ўзгартира олади.</li>
    </ul>
</div>

<div class="card">
    <h2>IT соҳасидаги қўлланилиши</h2>
    <p>Python турли соҳаларда қўлланилади:</p>
    <ul>
        <li><strong>Веб-ишланма</strong>: Django, Flask каби фреймворклар орқали веб-сайтлар ва веб-иловалар яратиш.</li>
        <li><strong>Машина ўрганиши (Machine Learning)</strong>: TensorFlow, Scikit-learn, Keras каби китобхоналар ёрдамида сунъий интеллект алгоритмларини ишлаб чиқиш.</li>
        <li><strong>Маълумотлар таҳлили</strong>: Pandas, NumPy, Matplotlib каби воситалар орқали маълумотларни таҳлил қилиш.</li>
        <li><strong>Автоматлаштириш</strong>: Selenium, PyAutoGUI каби воситалар ёрдамида вазифаларни автоматлаштириш.</li>
        <li><strong>Ойна дастурлари</strong>: Tkinter, PyQt каби график интерфейс иловаларини яратиш.</li>
    </ul>
</div>
